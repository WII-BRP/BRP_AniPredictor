```python
import os
import shutil
import pandas as pd
```

```python
dest_dir = r"G:\Trial\1"
```

```python
sub_dirs = []
for dirpath, dirnames, filenames in os.walk(dest_dir):
    if dirpath.split("\\")[-1]  in ["Cropped_images"]:
        sub_dirs.append(dirpath)
        
if sub_dirs == []:
    sub_dirs.append(dest_dir)

for i in sub_dirs:
    print(i)
```

```python
shutil.make_archive(i,format="zip",root_dir=i)
shutil.rmtree(i)
```

```python
predictions = pd.read_csv(os.path.join(dest_dir,"predictions.csv"))
predictions["Species_pred"].fillna(predictions["Order_pred"], inplace=True)
predictions["Species_pred_prob"].fillna(predictions["Order_pred_prob"], inplace=True)
predictions.drop(columns=["Species_dir","Species_level_path"], inplace = True)
predictions["file"] = predictions["Filepath"].apply(lambda x: x.split("\\")[-1])
predictions["bbox"] = predictions["Detection_bbox"]
predictions["category"] = predictions["Species_pred"]
predictions["conf"] = predictions["Species_pred_prob"]
predictions.tail()
```

```python
predictions.to_json(os.path.join(dest_dir,"predictions.json"), orient="records", indent = 4)
```
